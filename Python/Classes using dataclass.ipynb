# Traditional Class Structure
--- 

## for OOP

```python
class Dog:
    def __init__(self, name):
        self.name = name
        self.legs = 4
    
    def speak(self):
        print(self.name + ' says: Bark!')
        
```

```python
my_dog = Dog('Rover')
another_dog = Dog('Fluffy')
```

```python
my_dog.speak()
```

```python
another_dog.speak()
```

## Simplified Class

Using the dataclasses library

```python
from dataclasses import dataclass

@dataclass
class Dog:
    name: str = 'Puppy'
    legs: int = 4

    def speak(self):
        if self.name.lower() == ('monte'):
            print(self.name + " goes... WHOOOOOO WHOOOOOOO")  
        else:
            print(self.name + ' says: Bark!')
```

```python
nigel = Dog('Nigel')
monte = Dog('Monte')
luna = Dog('Luna')
baby = Dog()
```

```python
print(baby.name)
print(baby.legs)
baby.name = 'Thor'
print(baby.name)
```
